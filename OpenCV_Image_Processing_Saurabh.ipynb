# Image Processing using OpenCV 

```python
import cv2 as cv
import urllib.request  
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
```

### Downloading image from the url

```python
#Downloading image from url
urllib.request.urlretrieve("https://assets.myntassets.com/h_640,q_90,w_480/v1/assets/images/2354027/2018/1/16/11516097831111-MsTaken-Women-Navy-Blue-Striped-Top-4731516097830960-1.jpg", "image.jpg")
```

### Loading image from the url

```python
req = urllib.request.urlopen('https://assets.myntassets.com/h_640,q_90,w_480/v1/assets/images/2354027/2018/1/16/11516097831111-MsTaken-Women-Navy-Blue-Striped-Top-4731516097830960-1.jpg')
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
img = cv.imdecode(arr, -1) 
image = cv.cvtColor(img, cv.COLOR_BGR2RGB) 
plt.imshow(image) 
```

### Fetching image height & width


```python
print("Image_Height: %d" % img.shape[0])  
print("Image_Width: %d" % img.shape[1])  
```

### Resizing the image

```python
image_height = 224
image_width = 224
resized = cv.resize(image, (image_height, image_width))  
```

```python
#Displaying resized image using matplotlib
plt.imshow(resized) 
```

### Cropping the image

```python
cropped = image[115:579, 60:420]  
plt.imshow(cropped) 
```

### Saving the cropped image

```python
cv.imwrite("cropped_image.jpg", cv.cvtColor(cropped, cv.COLOR_RGB2BGR))   
```
