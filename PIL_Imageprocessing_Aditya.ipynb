```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
```

### Loading Image

```python
img=Image.open('shopin-image.jpg')
```

```python
from __future__ import print_function
print(img.format, img.size, img.mode)
```

```python
img.show()
```

```python
plt.imshow(img)
```

### Resizing Image

```python
img_resize=img.resize((224,224))
```

```python
img_resize.show()
```

```python
plt.imshow(img_resize)
```

### Cropping Image

```python
box=(60, 115, 420, 579)
img_crop=img.crop(box)
plt.imshow(img_crop)
```

### Saving image

```python
img_crop.save('img_crop_pil.jpg')
```
