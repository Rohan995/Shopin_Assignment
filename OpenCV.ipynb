```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import urllib
```

```python
##Downloading the image from the url

urllib.urlretrieve('https://assets.myntassets.com/h_640,q_90,w_480/v1/assets/images/2354027/2018/1/16/11516097831111-MsTaken-Women-Navy-Blue-Striped-Top-4731516097830960-1.jpg','Test_image.jpg')
```

```python
##Opening the image using the url

req = urllib.urlopen('https://assets.myntassets.com/h_640,q_90,w_480/v1/assets/images/2354027/2018/1/16/11516097831111-MsTaken-Women-Navy-Blue-Striped-Top-4731516097830960-1.jpg')
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
img = cv2.imdecode(arr, -1) # 'Load it as it is'

cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##Dimensions of the image
#Height, width , channel
img.shape
```

```python
##resizing the image 

resized_img = cv2.resize(img,(224,224))
resized_img.shape
```

```python
##displaying the resized image
rgb = cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.show()

# cv2.imshow('resized_img', resized_img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
##croping the image and displaying it

cropped_img = img[115:579,60:420]
cv2.imshow('cropped_img', cropped_img)
cv2.waitKey(0)
cv2.destroyAllWindows() 
```
