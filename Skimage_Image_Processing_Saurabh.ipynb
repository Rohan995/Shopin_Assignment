```python
img = skimage.io.imread("https://assets.myntassets.com/h_640,q_90,w_480/v1/assets/images/2354027/2018/1/16/11516097831111-MsTaken-Women-Navy-Blue-Striped-Top-4731516097830960-1.jpg") 
```

### Displaying image 

```python
plt.imshow(img) 
```

### Fetching image height & width 

```python
print("Image_Height: %d" % img.shape[0])  
print("Image_Width: %d" % img.shape[1])  
```

### Resizing the image 

```python
resized_image = resize(img, (224,224)) 
```

```python
plt.imshow(resized_image) 
```

### Cropping the image

```python
cropped_skimage = img[115:579, 60:420] 
```

```python
plt.imshow(cropped_skimage) 
```

### Saving the cropped image 

```python
skimage.io.imsave("cropped_skimage.jpg", cropped_skimage) 
```
