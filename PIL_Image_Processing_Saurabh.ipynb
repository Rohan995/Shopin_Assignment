# Image Processing using PIL 

```python
from PIL import Image
import urllib.request 
import io
```

### Downloading image from the url 

```python
urllib.request.urlretrieve("https://assets.myntassets.com/h_640,q_90,w_480/v1/assets/images/2354027/2018/1/16/11516097831111-MsTaken-Women-Navy-Blue-Striped-Top-4731516097830960-1.jpg", "image.jpg") 
```

### Loading image from the url

```python
ur = urllib.request.urlopen("https://assets.myntassets.com/h_640,q_90,w_480/v1/assets/images/2354027/2018/1/16/11516097831111-MsTaken-Women-Navy-Blue-Striped-Top-4731516097830960-1.jpg")
image_file = io.BytesIO(ur.read()) 
img = Image.open(image_file)   
```

### Displaying the image

```python
img
```

### Fetching image height & width

```python
print("Image_Height: %d" % img.size[0])  
print("Image_Width: %d" % img.size[1])  
```

### Resizing the image

```python
resized_image = img.resize((224,224)) 
```

```python
resized_image 
```

```python
print(resized_image.size) 
```

### Cropping the image

```python
dim = (60, 115, 420, 579)
cropped_pil = img.crop(dim) 
```

```python
cropped_pil 
```

### Saving the cropped image

```python
cropped_pil.save("cropped_pil.jpg") 
```
