```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import urllib.request
```

### Downloading Image

```python
urllib.request.urlretrieve('https://assets.myntassets.com/h_640,q_90,w_480/v1/assets/images/2354027/2018/1/16/11516097831111-MsTaken-Women-Navy-Blue-Striped-Top-4731516097830960-1.jpg','shopin-image.jpg')
```

```python
url='https://assets.myntassets.com/h_640,q_90,w_480/v1/assets/images/2354027/2018/1/16/11516097831111-MsTaken-Women-Navy-Blue-Striped-Top-4731516097830960-1.jpg'
```

### Loading image from URL

```python
res=urllib.request.urlopen(url)
```

```python
image = np.asarray(bytearray(res.read()), dtype="uint8")
```

```python
image = cv2.imdecode(image, cv2.IMREAD_COLOR)
```

### Showing image in RGB

```python
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

### Height and width of Image

```python
image.shape
```

### Resizing Image

```python
image_resize=cv2.resize(image,(224,224))
```

```python
plt.imshow(image_resize)
```

### Cropping Image

```python
pts1=np.float32([[0,0],[0,640],[480,0],[480,640]])
pts2=np.float32([[60,115],[60,579],[420,115],[420,579]])
M=cv2.getPerspectiveTransform(pts2,pts1)
crop_img=cv2.warpPerspective(image,M,(480,640))
plt.imshow(crop_img)
```

```python
pts3=np.float32([[0,0],[0,640],[480,0],[480,640]])
pts4=np.float32([[60,115],[60,579],[420,115],[420,579]])
M=cv2.getPerspectiveTransform(pts4,pts3)
crop_img1=cv2.warpPerspective(image,M,(420,579))
plt.imshow(crop_img1)
```

```python
crop_img2=image[115:579,60:420]
plt.imshow(crop_img2)
```

### Image saved in BGR

```python
cv2.imwrite('crop-img.jpg',crop_img)
cv2.imwrite('crop-img1.jpg',crop_img1)
```

### Image saved in RGB

```python
crop_imgRGB=cv2.cvtColor(crop_img,cv2.COLOR_RGB2BGR)
cv2.imwrite('crop-imgRGB.jpg',crop_imgRGB)
```

```python
crop_img1RGB=cv2.cvtColor(crop_img1,cv2.COLOR_RGB2BGR)
cv2.imwrite('crop-img1RGB.jpg',crop_img1RGB)
```

```python
crop_img2RGB=cv2.cvtColor(crop_img2,cv2.COLOR_RGB2BGR)
cv2.imwrite('crop-img1RGB.jpg',crop_img2RGB)
```
